### The columns Uk, England, Scotland, and Wales have numeric values stored as strings with commas. This prevents numerical operations, plotting or calculations directly on these columns. So we are identifying columns for conversion where a list is created that includes affected columns. The method `str.replace(',' , '')` removes commas from these columns, converting the numeric values into plain strings. Additionally after removing commas we convert the values to float and then into integer. This is necessary for enabling numerical operations, ensure data intrgrity and avoid errors in visualization. 

```python
# Remove commas and convert specified columns to numeric
column_convert = ['UK', 'England', 'Scotland', 'Wales']
for col in column_convert:
    data[col] = data[col].astype(str).str.replace(',', '').astype(float).astype(int)
    
# Preview of the columns converted to int 
print(data.head())
```

* Here we use `.str.strip()` which removes leading and trailing whitespace from each entry in the column

## Focusing on specific measure type for our analysis

### Household Waste Generated `(Arisings)`
* We are flitering the dataset to include only rows where column `Measure` equals `Arisings`(waste generated) and result is stored in `arising_data`.
* The filtered data is grouped by `year` column. For each year, the country columns are summed up. And the results are reset into a new DataFrame`arising_by_year` with the year as a separate column.
* The country columns are summed across all years to get the total waste generated for each country. The results are reshaped into a DataFrame with columns renamed to `Country` and `Total Arisings`.
 


### Insights 
England has generated the majority of the total waste that is `288,887thousand tonnes which is approximately 83% of the UK's total. Scotland`(30,804) and Wales`(16,516)` contributes less but still plays a significant role. Northern Ireland`(10,716)` has the smallest total waste generated. So, overall UK has generated `346,972` thousand tonnes over the period of time

### Key Insights
* `England` contibutes the majority of waste generated, signficantly surpassing the other regions
* While there are year-to-year fluctuions, UK waste generation remain fairly consistent over time, with minor peaks `(eg. 2021 at 27,650)`and declines `(eg. 2022 at 25,691)`.
* `Northern Ireland` contributes the least amount of waste generated, likely refelecting its smaller population and economic activity compared to other regions. 

```python
# Filtering data for a specific measure type 'arisings'
arising_data = data[data['Measure (thousand tonnes)'] == 'arisings']

# Grouping and aggregating data by year for 'Arisings'
arising_by_year = arising_data.groupby('Year')[['UK', 'England', 'NI', 'Scotland', 'Wales']].sum().reset_index()
print("Total 'Arisings' by Year:\n", arising_by_year)

# Aggregating waste data by country for 'Arisings'
arising_by_country = arising_data[['UK', 'England', 'NI', 'Scotland', 'Wales']].sum().reset_index()
arising_by_country.columns = ['Country', 'Total Arisings']
print("Total 'Arisings' by Country:\n", arising_by_country)
```

## Feature Engineering
### In feature engineering we took real values for the population DataFrame: For population of `UK` (https://www.macrotrends.net/global-metrics/countries/GBR/united-kingdom/population), For population of `England` (https://www.statista.com/statistics/975956/population-of-england/) , For population of `Northern Ireland` (https://www.statista.com/statistics/383733/northern-ireland-population-timeline-uk/) , For population of `Scotland` (https://www.scotlandscensus.gov.uk/2022-results/scotland-s-census-2022-rounded-population-estimates/#:~:text=On%20Census%20Day%2C%2020%20March,grew%20by%20233%2C400%20(4.6%25).&text=The%20other%20UK%20censuses%20showed,female%20and%2048.5%25%20was%20male.) , For population of `Wales` (https://www.gov.wales/mid-year-estimates-population-2022) these values are of the year `2022` 
The `waste per capita` and `recycling efficiency` metrics for UK and its regions is calculated over the years. Population data is used for calculations.

### Caluclations
* Waste per capita for each region is calculated using the formula `Waste Per Capita = Total Waste/Population` 
* Recycling Efficiency for each region is calculated using the formula `Recycling Efficiency(%) = Recycled Waste / Total Waste * 100`

### Insights 
Waste per capita
  * Waste per capita is relatively stable across years and regions, with little varaiations.
  * `Wales` and `Scotland` tend to generate higher waste per capita compared to England and Northern Ireland.

Recycling Efficiency
  * `Wales` has the highest recycling efficiency by achieving `56.01%` of recycling rate in `2021`, which outperforms other regions.
  * `Scotland` shows steady improvement, increasing from `32.50% (2010) to 41.79% (2022)`.
  * `England` shows steady efficiency around `43-44%` and `Northern Ireland` has improved reaching `49.26%(2022)`.
  * Overall UK's recycling efficiency has steadily improved from `40.36%(2010)` to `43.17%(2022)`, which reflects progress in waste management.

### Overveiw
